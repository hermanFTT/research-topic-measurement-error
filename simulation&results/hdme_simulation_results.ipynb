# On the implementation of correction methods to mitigate measurement error in high-dimensional data: Results obtained after simulations.

## Simulation setup:


*   p=200 ( number of covariate)
*   n=5000 ( observations)
*   sim=100 ( number of simulations)
*   under additive error assumption

## Process
Two different situations


1.  the number of non zero coefficients (relevant variables) is s=5. (result1)


*   train the models overall number of simulations and  record the MSE, PE (prediction error), $L_1$ error, number selected variables and number of correct selected variables.
*  get the mean values of all these 'metric' over the number of simulations


2.   The number of all zero coefficients is " s=10": follow the same workflow (result2)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names=['mse','l1_error','pe', 'num_var_selected','num_correct_var_selected']
```

#SIMULATION RESULTS FOR  " s=5 " (active set) NON ZERO COEFFICIENTS

```python
True_lasso_result1=pd.read_csv('True_lasso_result1.csv',header=None,names=column_names)
r=True_lasso_result1.fillna(0)
r=r.astype({ 'num_var_selected': int,'num_correct_var_selected': int})
True_lasso_result1=r.mask(r==0,r.mean(),axis=1)

mean_val_1=True_lasso_result1.mean()
print('mean vaues :\n', mean_val_1)
print(' ')
std_val_1=True_lasso_result1.std()
print('standard deviation values:\n', std_val_1)

True_lasso_result1
```

### True Dantzig results ( Dantzig Selector model)

```python
True_dantzig_result1=pd.read_csv('True_dantzig_result1.csv',header=None,names=column_names)
r=True_dantzig_result1.fillna(0)
r=r.astype({ 'num_var_selected': int,'num_correct_var_selected': int})
True_dantzig_result1=r.mask(r==0,r.mean(),axis=1)

mean_val_6=True_dantzig_result1.mean()
print('mean vaues :\n', mean_val_6)
print(' ')
std_val_6=True_dantzig_result1.std()
print('standard deviation values:\n', std_val_6)

True_dantzig_result1
```

### Naive lasso results

```python
Naive_lasso_result1=pd.read_csv('Naive_lasso_result.csv',header=None,names=column_names)
r=Naive_lasso_result1.fillna(0)
r=r.astype({ 'num_var_selected': int,'num_correct_var_selected': int})
Naive_lasso_result1=r.mask(r==0,r.mean(),axis=1)

mean_val_2=Naive_lasso_result1.mean()
print('mean vaues :\n', mean_val_2)
print(' ')
std_val_2=Naive_lasso_result1.std()
print('standard deviation values:\n', std_val_2)

Naive_lasso_result1
```

### Corrected Lasso Results (Non Convex Lasso)

```python
correct_lasso_result1=pd.read_csv('correct_lasso_result1.csv',header=None,names=column_names)
r=correct_lasso_result1.fillna(0)
r=r.astype({ 'num_var_selected': int,'num_correct_var_selected': int})
correct_lasso_result1=r.mask(r==0,r.mean(),axis=1)

mean_val_3=correct_lasso_result1.mean()
print('mean vaues :\n', mean_val_3)
print(' ')
std_val_3=correct_lasso_result1.std()
print('standard deviation values:\n', std_val_3)

correct_lasso_result1
```

### CoCoLasso results (Convex conditional Lasso)

```python
Cocolasso_result1=pd.read_csv('Cocolasso_result1.csv',header=None,names=column_names)
r=Cocolasso_result1.fillna(0)
r=r.astype({ 'num_var_selected': int,'num_correct_var_selected': int})
Cocolasso_result1=r.mask(r==0,r.mean(),axis=1)

mean_val_4=Cocolasso_result1.mean()
print('mean vaues :\n', mean_val_4)
print(' ')
std_val_4=Cocolasso_result1.std()
print('standard deviation values:\n', std_val_4)

Cocolasso_result1
```

```python
MUS_result1=pd.read_csv('MUS_result1.csv',header=None,names=column_names)
r=MUS_result1.fillna(0)
r=r.astype({ 'num_var_selected': int,'num_correct_var_selected': int})
MUS_result1=r.mask(r==0,r.mean(),axis=1)

mean_val_5=MUS_result1.mean()
print('mean vaues :\n', mean_val_5)
print(' ')
std_val_5=MUS_result1.std()
print('standard deviation values:\n', std_val_5)

MUS_result1
```

Summary mean values.

```python
mean_val=pd.DataFrame({'True_lasso':mean_val_1 ,
                      'Naive_lasso':mean_val_2  ,
                      'True_Dantzig': mean_val_6,
                      'Non_convex_lasso': mean_val_3 ,
                      'Coco_lasso': mean_val_4 ,
                      'MUS':mean_val_5})

print('mean values for each model\n')
pd.concat([mean_val.iloc[0:3],mean_val.iloc[-2:].round(decimals=0)])
```

Summary standard deviation

```python
std_val=pd.DataFrame({'True_lasso':std_val_1 ,
                      'Naive_lasso':std_val_2  ,
                      'True_Dantzig': std_val_6,
                      'Non_convex_lasso': std_val_3 ,
                      'Coco_lasso': std_val_4 ,
                      'MUS':std_val_5})

print('standard values for each model\n')
pd.concat([std_val.iloc[0:3],std_val.iloc[-2:].round(decimals=0)])
```

```python
X=np.arange(80)
```

```python
plt.figure( figsize=(15,5))
sns.set_style("whitegrid")
plt.plot(X,True_lasso_result1['num_var_selected'], marker='*', label='True_lasso')
plt.plot(X, Naive_lasso_result1['num_var_selected'], marker='o', label='Naive_lasso',alpha=0.3,color='red')
plt.plot(X, correct_lasso_result1['num_var_selected'], marker='+', label='NCL',alpha=0.8,color='green')
plt.plot(X, Cocolasso_result1['num_var_selected'], marker='.', label='coco_lasso',alpha=0.4,color='magenta')
plt.plot(X, MUS_result1['num_var_selected'], marker='.', label='MUS',alpha=0.6,color='black')

plt.title('selected variables trend for each model')
plt.xlabel('simulation')
plt.ylabel('number of selected variables')
plt.legend(title='Legend', loc='upper center')
# Show plot
plt.show()

```

```python
plt.figure( figsize=(15,5))
sns.set_style("whitegrid")
plt.plot(X,True_lasso_result1['mse'], marker='*', label='True_lasso')
plt.plot(X, Naive_lasso_result1['mse'], marker='o', label='Naive_lasso',alpha=0.4,color='red')
plt.plot(X, correct_lasso_result1['mse'], marker='+', label='NCL',alpha=0.3,color='green')
plt.plot(X, Cocolasso_result1['mse'], marker='.', label='coco_lasso',alpha=0.7,color='magenta')
plt.plot(X, MUS_result1['mse'], marker='.', label='MUS',alpha=0.4,color='black')

plt.title(' MSE of models over simulations')
plt.xlabel('simulation')
plt.ylabel('mse ')
plt.legend(title='Legend', loc='upper center')
# Show plot
plt.show()
```
